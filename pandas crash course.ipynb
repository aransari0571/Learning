```python
vis = sns.distplot(df['InternetUsers'])
plt.show()
```

```python
vis1 = sns.displot(df['InternetUsers'])
plt.show()
```

```python
vis2 = sns.distplot(df['InternetUsers'], bins =15)
plt.show()
```

```python
vis3 = sns.boxplot(data = df, x="IncomeGroup", y='BirthRate')
plt.show(vis3)
```

```python
vis4 = sns.lmplot(data = df, x="InternetUsers", y='BirthRate')
plt.show(vis4)
```

```python
vis4 = sns.lmplot(data = df, x="InternetUsers", y='BirthRate', fit_reg=False)
plt.show(vis4)
```

```python
vis5 = sns.lmplot(data = df, x="InternetUsers", y='BirthRate', fit_reg=False, hue='IncomeGroup')# hue is parameter f color
plt.show(vis5)
```

```python
vis6 = sns.lmplot(data = df, x="InternetUsers", y='BirthRate', fit_reg=True, hue='IncomeGroup')# hue is parameter f color
plt.show(vis6)
```
